```python
#this is for problem 1:
#part a: calculate P(O|lam)
A = np.array([[0.7,0.3],[0.4,0.6]]).T
B = np.array([[0.1,0.4,0.5],[0.7,0.2,0.1]]).T
pi = np.array([0.0,1.0]).T
lam = np.array([A,B,pi])

obs = np.array([1,0,2])

def P_O_lam(A,B,pi, observations):
    T = len(observations)
    N = np.shape(A)[0] #this is the number of rows in A
    tot = 0 #running total probability...will be updated
    output = []
    
    for i in it.product(xrange(N), repeat = T):
        P = pi[i[0]]*B[observations[0],i[0]]
        
        for j in xrange(1, T):
            P *= A[i[j], i[j - 1]] * B[observations[j], i[j]]
        tot += P
        output.append((i, P))
        
    return output, tot
part_a = P_O_lam(A,B,pi,obs)
print "this is the output: \n", part_a, "\n this is the total probability:\n", part_a[1]
    
print "\n we now compute the alpha pass, this is part b \n"

def alpha_pass(A,B,pi,observations, return_all = False):
    if return_all:
        alpha_list = []
    for t in xrange(len(observations)):
        j = int(observations[t])
       
        if t == 0:
            alpha = pi * B[j, :]
        else:
            alpha = A.dot(alpha) * B[j, :]
            
        if return_all:
            alpha_list.append(alpha)
        
    if return_all:
        return alpha, alpha_list
    else:
        return alpha
print "this is my alpha pass:\n",alpha_pass(A,B,pi,obs, return_all = True)


    
```

```python
#in order to scale things I'll need a beta pass:
def beta_pass(A, B, pi, Ob, return_all = False):
    if return_all:
        beta_list = []
        
    for t in xrange(len(Ob)):
        j = int(Ob[-t])
        
        if t == 0:
            beta = np.ones_like(pi)
        else:
            beta = A.T.dot(beta * B[j, :])
            
        if return_all:
            beta_list.append(beta)
    
    if return_all:
        return beta, beta_list
    else:
        return beta
    
    
def scale_alpha(A, B, pi, Ob):
    T = len(Ob)
    N = len(pi)
    c = np.zeros(T)

    alpha = np.zeros((T, N))
    alpha[0]  = pi * B[Ob[0]]
    c[0] = 1. / alpha[0].sum()
    alpha[0] *= c[0]

    for t in xrange(1, T):
        alpha[t] = np.dot(A.T, alpha[t - 1]) * B[Ob[t]]
        c[t] = 1. / alpha[t].sum()
        alpha[t] *= c[t]

    return alpha, c #we want to return C because it will be used in the beta scaling

def scale_beta(A, B, pi, Ob, c):
    T = len(Ob)
    N = len(pi)
    d = c

    beta = np.zeros((T, N))
    beta[-1] = np.ones(N)
    d[-1] = 1. / beta[-1].sum()
    beta[-1] *= d[-1]

    for t in range(0, T - 1)[::-1]:
        beta[t] = np.dot(A.T, beta[t + 1] * B[Ob[t + 1]])
        d[t] = 1. / beta[t].sum()
        beta[t] *= d[t]

    return beta

def stoch(S, M):
    A = np.random.rand(S, S)
    A /= np.hstack(A.sum(axis = 0))
    
    B = np.random.rand(M, S)
    B /= np.hstack(B.sum(axis = 0))

    pi = np.random.rand(S)
    pi /= pi.sum()

    return A, B, pi

def clean(datafile):
    f = open(datafile)
    data = f.read()[:50000]
    Ob = list(data)
    T = len(Ob)
    i = 0

    while i != T:
        pos = ord(Ob[i])
        if pos == 32:
            Ob[i] = 0

        elif pos >= 65 and pos <= 90:
            Ob[i] = pos - 64

        elif pos >= 97 and pos <= 122:
            Ob[i] = pos - 96

        else:
            del Ob[i]
            T -= 1
            i -= 1
        i += 1
    return Ob, T


def prob9(datafile, S, maxiters = 10):
    M = 27
    iters = 0
    O, T = clean(datafile)
    A,B,pi = stoch(S,M)
    logprob = -np.inf
    oldLogProb = -np.inf
    diff = 1
    while(iters < maxiters and diff > 0): #this keeps updating our model
        alpha, c = scale_alpha(A,B,pi,O)
        beta = scale_beta(A,B,pi,O,c)

        dig = np.zeros((T,S,S))
        gamma = np.zeros((T,S))
        for t in xrange(T-1):
            for i in xrange(S):
                for j in xrange(S):
                    dig[t,i,j] = alpha[t,i]*A[j,i]*B[O[t+1],j]*beta[t+1,j]
            dig[t] = dig[t]/dig[t].sum()

        gamma = dig.sum(axis =1)
        pi = gamma[0]
        for i in range(S):
            for j in range(S):
                A[i,j] = dig[:,i,j].sum()/gamma[:,i].sum()
        for i in range(S):
            for j in range(M):
                numer = 0
                for t in range(T):
                    if O[t] == j : numer += gamma[t,i]
                denom = gamma[:,i].sum()
                B[j,i] = numer/denom
        oldLogProb = logprob

        logprob = -c.sum()
        dif = logprob*oldLogProb
        iters +=1
    return pi, A, B.argmax(axis=1)

datafile = 'brown.txt'
print prob9(datafile,2,10)

print "trying to do it with 3 states\n"
print prob9(datafile,3,10)

print "trying to do it with 4 states\n"
print prob9(datafile,4,10)
```
